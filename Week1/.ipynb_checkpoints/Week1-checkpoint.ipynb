## Part 1
The given algorithm for computing fibonacci numbers is implemented in C below. The contents of the cell below are written into the file __fib.c__ whenever it is run.

```python
%%writefile fib.c
#include<stdlib.h>
#include<stdio.h>

int main(){
    int n=1, nold=1, new=0,k;
    printf("1 %d\n",n);
    printf("2 %d\n",nold);
    for(k=3;k<=10;k++){
        new = n+nold;
        nold=n;
        n=new;
        printf("%d %d\n",k,new);
    }
}
```

Terminal commands are run using the __!__ prefix.

```python
!gcc fib.c -o fib.out
```

```python
!./fib.out
```

Now, the python implementation:

```python
def fib():
    n=1
    nold=1
    new=0
    for k in range(3,11,1):
        new=n+nold
        nold=n
        n=new
        print k,new
```

```python
fib()
```

The python implementation:

```python
def listProgram():
    import math
    n=[0.2]
    for k in range(1,1000):
        t = (n[k-1]+math.pi)*100
        n.append(t-int(t))
    
    for i in n:
        print("%.4f"%i)
```

```python
listProgram()
```

## Part 3
The third part is implemented below:

```python
def freq():
    import re
    
    with open('the-hound-of-the-baskervilles.txt', 'r') as f:
        contents = f.read()
    
    d = {}
    words = re.split("[. !?\-\n\[\],\"\']", contents)
    
    for word in words:
        if len(word.strip())==0:continue
        d[word] = d.get(word,0)+1
            
    for key in sorted(d):
        print key,d[key]
```

```python
freq()
```
